#### How can I turn the messiness of the cnvs and calls into a call that is useful?

```python
df_TP53.describe()
```

A meaningful call would probably involve giving a weight to the cnv.<br>
According to the instructions, a cnv of 1-7 is relatively neutral, and should be weighted as such<br>
CNV of 0 is a total absence of copies, and should carry the maximum negative weight<br>
CNV of 8-20(20 being the max found in all of the patients) would have ascending weight.

```python
TP53_PAT1012 = df_TP53.loc[df_TP53.orderid == 'PAT1012'] 
TP53_PAT1012
```

```python
TP53_1 = df_TP53.loc[df_TP53.probe == 1] 
TP53_1.describe()
```

```python
TP53_PAT1012mod = TP53_PAT1012
TP53_PAT1012mod 
```

Here I make a function with arbitrary weights that follow the logic stated earlier.<br>
The mean and evaluation are not made in this function because there may be some pattern in the location or frequency of the weighted values that could potentially offer an insight(purely hypothetical)

```python

#Made up weights for each of the cnvs based on 0 being a total deletion, and the max based on the max across all patients,
#cnv of 1-7 is just weighted as 0(neutral)

def convert_to_weight(cnvs):
    cnvs['cnv'].replace([0], -1000,inplace=True)
    cnvs['cnv'].replace([1,2,3,4,5,6,7], 0, inplace=True)
    cnvs['cnv'].replace([8,9,10], 100,inplace=True)
    cnvs['cnv'].replace([11,12,13,14],300,inplace=True)
    cnvs['cnv'].replace([15,16,17,18,19,20], 1000,inplace=True)
    
    return cnvs     
    
```

```python
convert_to_weight(TP53_PAT1012mod)
```

An arbitrary function that performs some calculations and returns an evaluation on the significant of the cnvs

```python
def evaluate_weight(mean):
    x = mean/200
    print(x)
    if int(round(x)) == 0:
        return 'neutral'
    elif abs(int(round(x))) >=2:
        return 'significant'
    else:
        return 'potentially significant'
    
```

```python
mean1 = TP53_PAT1012mod.cnv.mean() 
evaluate_weight(mean1)
```
